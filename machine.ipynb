```python
# Solución Optimizada para tu DataFrame datos_ml
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from catboost import CatBoostClassifier
from imblearn.over_sampling import SMOTE
import warnings
warnings.filterwarnings('ignore')

# PASO 1: ANÁLISIS DE TU DATOS_ML ACTUAL
def analizar_datos_ml(datos_ml):
    """Analizar el estado actual de datos_ml"""
    print("=== ANÁLISIS DE DATOS_ML ACTUAL ===")
    print(f"Forma: {datos_ml.shape}")
    print(f"Columnas: {list(datos_ml.columns)}")

    # Verificar el mapeo de la variable objetivo
    print(f"\nDistribución de estado_aprobación:")
    print(datos_ml['estado_aprobación'].value_counts().sort_index())

    # Estadísticas por clase
    for estado in [0, 1, 2]:  # Rejected, Pending, Approved
        subset = datos_ml[datos_ml['estado_aprobación'] == estado]
        estado_nombre = ['Rejected', 'Pending', 'Approved'][estado]
        print(f"\n{estado_nombre} (clase {estado}):")
        print(f"  Precio promedio: ${subset['precio_usd'].mean():.2f}")
        print(f"  Dosis promedio: {subset['dosis_mg'].mean():.0f}mg")
        print(f"  Año promedio: {subset['año_aprobación'].mean():.0f}")
```

```python
# PASO 2: CREAR CARACTERÍSTICAS MEJORADAS PARA DATOS_ML
def mejorar_datos_ml(datos_ml):
    """Agregar características discriminativas a datos_ml"""
    print("\n=== MEJORANDO DATOS_ML CON NUEVAS CARACTERÍSTICAS ===")

    df = datos_ml.copy()

    # 1. Características de efectos secundarios (análisis de texto)
    df['num_efectos_secundarios'] = df['efectos_secundarios'].str.count(',') + 1
    df['tiene_nausea'] = df['efectos_secundarios'].str.contains('Nausea', case=False).astype(int)
    df['tiene_mareos'] = df['efectos_secundarios'].str.contains('Dizziness', case=False).astype(int)
    df['tiene_fatiga'] = df['efectos_secundarios'].str.contains('Fatigue', case=False).astype(int)
    df['tiene_dolor_cabeza'] = df['efectos_secundarios'].str.contains('Headache', case=False).astype(int)
    df['tiene_vision_borrosa'] = df['efectos_secundarios'].str.contains('Blurred vision', case=False).astype(int)
    df['tiene_boca_seca'] = df['efectos_secundarios'].str.contains('Dry mouth', case=False).astype(int)
    df['tiene_insomnio'] = df['efectos_secundarios'].str.contains('Insomnia', case=False).astype(int)

    # Efectos agrupados
    df['efectos_neurologicos'] = (df['tiene_mareos'] + df['tiene_dolor_cabeza'] +
                                 df['tiene_vision_borrosa']).clip(0, 1)
    df['efectos_gastrointestinales'] = df['efectos_secundarios'].str.contains(
        'Nausea|Diarrhea|Constipation', case=False).astype(int)
    df['efectos_graves'] = (df['tiene_vision_borrosa'] + df['tiene_boca_seca'] +
                           df['tiene_insomnio']).clip(0, 1)

    # 2. Características económicas y de dosis
    df['precio_por_mg'] = df['precio_usd'] / (df['dosis_mg'] + 1)
    df['es_medicamento_caro'] = (df['precio_usd'] > df['precio_usd'].quantile(0.7)).astype(int)
    df['es_medicamento_barato'] = (df['precio_usd'] < df['precio_usd'].quantile(0.3)).astype(int)
    df['es_dosis_alta'] = (df['dosis_mg'] > df['dosis_mg'].quantile(0.7)).astype(int)
    df['es_dosis_baja'] = (df['dosis_mg'] < df['dosis_mg'].quantile(0.3)).astype(int)
    df['es_dosis_extrema'] = (df['es_dosis_alta'] | df['es_dosis_baja']).astype(int)

    # 3. Características temporales
    df['años_desde_aprobacion'] = 2024 - df['año_aprobación']
    df['es_medicamento_antiguo'] = (df['año_aprobación'] < 2000).astype(int)
    df['es_medicamento_muy_antiguo'] = (df['año_aprobación'] < 1995).astype(int)
    df['es_medicamento_nuevo'] = (df['año_aprobación'] > 2015).astype(int)
    df['es_medicamento_muy_nuevo'] = (df['año_aprobación'] > 2020).astype(int)

    # 4. Características de gravedad
    df['gravedad_numerica'] = df['gravedad_efectos_secundarios'].map({
        'Mild': 1, 'Moderate': 2, 'Severe': 3
    })
    df['es_gravedad_severa'] = (df['gravedad_numerica'] == 3).astype(int)
    df['es_gravedad_leve'] = (df['gravedad_numerica'] == 1).astype(int)

    # 5. Score de riesgo combinado
    df['riesgo_score'] = df['gravedad_numerica'] * df['num_efectos_secundarios']
    df['riesgo_alto'] = (df['riesgo_score'] > df['riesgo_score'].quantile(0.75)).astype(int)
    df['riesgo_bajo'] = (df['riesgo_score'] < df['riesgo_score'].quantile(0.25)).astype(int)

    # 6. Características de fabricante
    fabricantes_grandes = ['Pfizer Inc.', 'Johnson & Johnson', 'Merck & Co.',
                          'Novartis AG', 'Roche Holding AG', 'AstraZeneca', 'Bayer AG']
    df['fabricante_grande'] = df['fabricante'].isin(fabricantes_grandes).astype(int)

    # 7. Características de clase de medicamento
    medicamentos_psiquiatricos = ['Antidepressant', 'Antipsychotic']
    medicamentos_antiinfecciosos = ['Antibiotic', 'Antifungal', 'Antiviral']
    medicamentos_dolor = ['Analgesic', 'Anti-inflammatory']

    df['es_psiquiatrico'] = df['clase_medicamento'].isin(medicamentos_psiquiatricos).astype(int)
    df['es_antiinfeccioso'] = df['clase_medicamento'].isin(medicamentos_antiinfecciosos).astype(int)
    df['es_para_dolor'] = df['clase_medicamento'].isin(medicamentos_dolor).astype(int)

    # 8. Características de contraindicaciones
    df['contraindicacion_embarazo'] = df['contraindicaciones'].str.contains(
        'Pregnancy', case=False).astype(int)
    df['contraindicacion_cardiaca'] = df['contraindicaciones'].str.contains(
        'Heart|Blood pressure', case=False).astype(int)
    df['contraindicacion_hepatica'] = df['contraindicaciones'].str.contains(
        'Liver', case=False).astype(int)
    df['contraindicacion_renal'] = df['contraindicaciones'].str.contains(
        'Kidney', case=False).astype(int)

    # 9. Características de advertencias
    df['advertencia_alcohol'] = df['advertencias'].str.contains('alcohol', case=False).astype(int)
    df['advertencia_maquinaria'] = df['advertencias'].str.contains('machinery', case=False).astype(int)
    df['advertencia_somnolencia'] = df['advertencias'].str.contains('drowsiness', case=False).astype(int)

    # 10. Interacciones clave (características que podrían ser discriminativas)
    df['precio_alto_riesgo_bajo'] = ((df['es_medicamento_caro'] == 1) &
                                    (df['riesgo_bajo'] == 1)).astype(int)
    df['precio_bajo_riesgo_alto'] = ((df['es_medicamento_barato'] == 1) &
                                    (df['riesgo_alto'] == 1)).astype(int)
    df['nuevo_y_caro'] = ((df['es_medicamento_nuevo'] == 1) &
                         (df['es_medicamento_caro'] == 1)).astype(int)
    df['antiguo_y_barato'] = ((df['es_medicamento_antiguo'] == 1) &
                             (df['es_medicamento_barato'] == 1)).astype(int)

    # 11. Patrones específicos observados (basado en análisis de confusión)
    df['patron_aprobacion_favorable'] = (
        (df['fabricante_grande'] == 1) &
        (df['es_gravedad_severa'] == 0) &
        (df['num_efectos_secundarios'] <= 2)
    ).astype(int)

    df['patron_rechazo_probable'] = (
        (df['es_gravedad_severa'] == 1) |
        (df['num_efectos_secundarios'] >= 4) |
        (df['precio_bajo_riesgo_alto'] == 1)
    ).astype(int)

    nuevas_caracteristicas = len(df.columns) - len(datos_ml.columns)
    print(f"Se crearon {nuevas_caracteristicas} nuevas características")
    print(f"Dataset final: {df.shape}")

    return df
```

```python
# PASO 3: MODELO CATBOOST MEJORADO ESPECÍFICO
def entrenar_modelo_mejorado(datos_ml_mejorado):
    """Entrenar modelo CatBoost optimizado con los datos mejorados"""
    print("\n=== ENTRENANDO MODELO CATBOOST MEJORADO ===")

    # Preparar X e y
    X = datos_ml_mejorado.drop(columns=['estado_aprobación'])
    y = datos_ml_mejorado['estado_aprobación']

    # Identificar características categóricas
    categorical_columns = ['fabricante', 'clase_medicamento', 'indicaciones',
                          'efectos_secundarios', 'vía_administración',
                          'contraindicaciones', 'advertencias',
                          'gravedad_efectos_secundarios']

    cat_features = []
    for i, col in enumerate(X.columns):
        if col in categorical_columns:
            cat_features.append(i)

    print(f"Características categóricas: {len(cat_features)}")
    print(f"Total características: {X.shape[1]}")

    # División de datos
    X_temp, X_test, y_temp, y_test = train_test_split(
        X, y, test_size=0.15, stratify=y, random_state=42
    )
    X_train, X_val, y_train, y_val = train_test_split(
        X_temp, y_temp, test_size=0.3, stratify=y_temp, random_state=5
    )

    print(f"Tamaños: Train={X_train.shape[0]}, Val={X_val.shape[0]}, Test={X_test.shape[0]}")

    # Aplicar SMOTE para balancear
    print("Aplicando SMOTE...")
    smote = SMOTE(random_state=42, k_neighbors=5)
    X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)

    print("Distribución original:")
    print(pd.Series(y_train).value_counts().sort_index())
    print("Distribución balanceada:")
    print(pd.Series(y_train_balanced).value_counts().sort_index())

    # Modelo CatBoost optimizado
    modelo = CatBoostClassifier(
        iterations=2000,
        depth=10,
        learning_rate=0.05,
        l2_leaf_reg=3,
        border_count=128,
        bagging_temperature=0.5,
        random_strength=0.3,
        loss_function='MultiClass',
        eval_metric='Accuracy',
        random_seed=42,
        verbose=400,
        early_stopping_rounds=150,
        use_best_model=True,
        class_weights=[1.0, 1.2, 1.0]  # Dar más peso a Pending
    )

    # Entrenar
    print("Entrenando modelo...")
    modelo.fit(
        X_train_balanced, y_train_balanced,
        cat_features=cat_features,
        eval_set=(X_val, y_val),
        plot=False
    )

    # Predicciones
    y_train_pred = modelo.predict(X_train_balanced)
    y_val_pred = modelo.predict(X_val)
    y_test_pred = modelo.predict(X_test)

    # Métricas
    train_acc = accuracy_score(y_train_balanced, y_train_pred)
    val_acc = accuracy_score(y_val, y_val_pred)
    test_acc = accuracy_score(y_test, y_test_pred)

    print(f"\nRESULTADOS:")
    print(f"Train Accuracy: {train_acc:.4f} ({train_acc*100:.1f}%)")
    print(f"Validation Accuracy: {val_acc:.4f} ({val_acc*100:.1f}%)")
    print(f"Test Accuracy: {test_acc:.4f} ({test_acc*100:.1f}%)")

    return modelo, X_test, y_test, y_test_pred, X

# PASO 4: EVALUACIÓN DETALLADA
def evaluacion_completa(modelo, X_test, y_test, y_test_pred, X):
    """Evaluación completa del modelo mejorado"""
    print(f"\n=== EVALUACIÓN COMPLETA ===")

    accuracy_final = accuracy_score(y_test, y_test_pred)
    print(f"ACCURACY FINAL: {accuracy_final:.4f} ({accuracy_final*100:.1f}%)")

    # Reporte de clasificación
    target_names = ['Rejected', 'Pending', 'Approved']
    print(f"\nREPORTE DE CLASIFICACIÓN:")
    print(classification_report(y_test, y_test_pred, target_names=target_names))

    # Matriz de confusión
    cm = confusion_matrix(y_test, y_test_pred)
    print(f"\nMATRIZ DE CONFUSIÓN:")
    print(cm)

    # Visualización
    plt.figure(figsize=(15, 10))

    # Matriz de confusión
    plt.subplot(2, 3, 1)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=target_names, yticklabels=target_names)
    plt.title('Matriz de Confusión')
    plt.ylabel('Real')
    plt.xlabel('Predicción')

    # Matriz normalizada
    plt.subplot(2, 3, 2)
    cm_norm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    sns.heatmap(cm_norm, annot=True, fmt='.2f', cmap='Blues',
                xticklabels=target_names, yticklabels=target_names)
    plt.title('Matriz Normalizada')
    plt.ylabel('Real')
    plt.xlabel('Predicción')

    # Importancia de características (Top 15)
    if hasattr(modelo, 'feature_importances_'):
        importancias = modelo.feature_importances_
        indices = np.argsort(importancias)[::-1][:15]

        plt.subplot(2, 3, 3)
        plt.barh(range(15), importancias[indices][::-1])
        plt.yticks(range(15), [X.columns[i] for i in indices[::-1]])
        plt.title('Top 15 Características')
        plt.xlabel('Importancia')

        plt.subplot(2, 3, 4)
        # Comparación de métricas por clase
        report = classification_report(y_test, y_test_pred, target_names=target_names, output_dict=True)
        metricas = ['precision', 'recall', 'f1-score']
        x_pos = np.arange(len(target_names))
        width = 0.25

        for i, metrica in enumerate(metricas):
            valores = [report[clase][metrica] for clase in target_names]
            plt.bar(x_pos + i*width, valores, width, label=metrica)

        plt.xlabel('Clases')
        plt.ylabel('Puntaje')
        plt.title('Métricas por Clase')
        plt.xticks(x_pos + width, target_names)
        plt.legend()
        plt.ylim(0, 1)

    # Análisis de errores
    plt.subplot(2, 3, 5)
    errores = y_test != y_test_pred
    error_por_clase = []
    for clase in range(3):
        mask_clase = y_test == clase
        errores_clase = errores[mask_clase].sum()
        total_clase = mask_clase.sum()
        tasa_error = errores_clase / total_clase if total_clase > 0 else 0
        error_por_clase.append(tasa_error)

    plt.bar(target_names, error_por_clase)
    plt.title('Tasa de Error por Clase')
    plt.ylabel('Tasa de Error')
    plt.ylim(0, 1)

    # Distribución de confianza
    plt.subplot(2, 3, 6)
    if hasattr(modelo, 'predict_proba'):
        probas = modelo.predict_proba(X_test)
        max_probas = np.max(probas, axis=1)
        plt.hist(max_probas, bins=20, alpha=0.7)
        plt.title('Distribución de Confianza')
        plt.xlabel('Máxima Probabilidad')
        plt.ylabel('Frecuencia')

    plt.tight_layout()
    plt.show()

    # Top características importantes
    if hasattr(modelo, 'feature_importances_'):
        print(f"\nTOP 10 CARACTERÍSTICAS MÁS IMPORTANTES:")
        for i in range(10):
            idx = indices[i]
            print(f"{i+1:2d}. {X.columns[idx]:30s} {importancias[idx]:.4f}")

    return accuracy_final

# FUNCIÓN PRINCIPAL
def mejorar_modelo_completo(datos_ml):
    """Función principal para mejorar tu modelo actual"""
    print("=== MEJORA COMPLETA DEL MODELO PARA DATOS_ML ===")

    # 1. Analizar datos actuales
    analizar_datos_ml(datos_ml)

    # 2. Crear características mejoradas
    datos_ml_mejorado = mejorar_datos_ml(datos_ml)

    # 3. Entrenar modelo mejorado
    modelo, X_test, y_test, y_test_pred, X = entrenar_modelo_mejorado(datos_ml_mejorado)

    # 4. Evaluación completa
    accuracy_final = evaluacion_completa(modelo, X_test, y_test, y_test_pred, X)

    print(f"\n{'='*60}")
    print(f"RESULTADO FINAL:")
    print(f"ACCURACY ANTERIOR: ~37%")
    print(f"ACCURACY NUEVO: {accuracy_final*100:.1f}%")
    print(f"MEJORA: {(accuracy_final - 0.37)*100:.1f} puntos porcentuales")
    print(f"{'='*60}")

    return modelo, datos_ml_mejorado

# EJECUTAR
print("""
=== INSTRUCCIONES ===

Para mejorar tu modelo actual, ejecuta:

modelo_mejorado, datos_ml_final = mejorar_modelo_completo(datos_ml)

Esto va a:
✅ Analizar tu datos_ml actual
✅ Crear ~40 características nuevas
✅ Entrenar CatBoost optimizado
✅ Aplicar SMOTE para balancear
✅ Mostrar evaluación completa

OBJETIVO: Superar 45-50% accuracy
""")
```
