```python
# biblioteca

import sklearn
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# hecer copia profunda del dataframe
import copy
datos_ml = copy.deepcopy(data)
```

```python
# eliminar columnas

datos_ml = datos_ml.drop(columns=['número_lote', 'fecha_caducidad'])
```

```python
datos_ml.head(7)
```

```python
# variables para machine

X = datos_ml.drop(columns=['estado_aprobación'])
y = datos_ml['estado_aprobación']
```

```python
X = pd.get_dummies(X)
```

```python
# dividir la prueba
X,X_test,y,y_test = train_test_split(X,y,test_size=0.15,stratify=y,random_state=42)
X_train,X_val,y_train,y_val = train_test_split(X,y,stratify=y,random_state=5)
```

```python
modelo = RandomForestClassifier(
    random_state=42,
    max_depth=4,
    class_weight='balanced'
    )
modelo.fit(X_train, y_train)
print(f'La precisión del modelo train es de : {modelo.score(X_train, y_train) * 100:.2f}%')
print(f'La precisión del modelo de validación es de : {modelo.score(X_val, y_val) * 100:.2f}%')
```

```python
y_pred = modelo.predict(X_test)
```

```python
# Evaluación
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nReporte de Clasificación:\n", classification_report(y_test, y_pred))
print("\nMatriz de Confusión:\n", confusion_matrix(y_test, y_pred))
```
